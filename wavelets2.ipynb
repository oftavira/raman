```python
# In this cell we import pywt, numpy, and matplotlib.pyplot
# We also define a signal x(t) that is the sum of two sinusoids

import numpy as np
import matplotlib.pyplot as plt
import pywt
import scipy.signal as sig
t = np.linspace(0, 1, 1000)
x = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*2*t)
coef, freqs = pywt.cwt(x, np.arange(1, 50), 'gaus1')

# After calculating the wavelet coefficients, we plot the coefficients
# with a large size to see the structure of the coefficients 

plt.figure(figsize=(15, 5))
plt.imshow(coef, aspect='auto', vmax=abs(coef).max(), vmin=-abs(coef).max())
```
